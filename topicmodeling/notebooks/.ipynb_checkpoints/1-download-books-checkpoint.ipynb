```python
import requests
import numpy as np
import re
import os
```

```python
BASEURL = r'http://www.gutenberg.org/cache/epub/%s/pg%s.txt'

bookids = [22523, 40087, 15896, 25341, 16588, 15524, 47248, 26930, 47269, 20335, 35075,
           20435, 13045, 29443, 29501, 29493, 30047, 24570, 6082, 26879, 793, 7947,
           23601, 33438, 10409, 30114, 35937, 27477, 25267, 28570, 6630, 15636, 
           35744, 8172, 16767, 26556, 25992, 27378, 39142, 58810]

start_regex = re.compile(r'\*\*\*[ ]*Start of [^\*]+\*\*\*', re.IGNORECASE)
end_regex = re.compile(r'\*\*\*[ ]*End of [^\*]+\*\*\*', re.IGNORECASE)

OUTPUT_FOLDER = 'books/wars_astronomy'
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
```

```python
for bookid in bookids:
    print('Download book %s, ....' % str(bookid),)
          
    url = BASEURL % (str(bookid), str(bookid))
    r = requests.get(url)
    booktxt = r.text
    
    m = start_regex.search(booktxt)
    if m is None:
        print('{} start regex missing'.format(bookid))
              
    sidx, eidx = m.span()
    
    booktxt = booktxt[eidx:]
    
    m = end_regex.search(booktxt)
    if m is None:
        print('{} end regex missing'.format(bookid))
        
    sidx, eidx = m.span()
    booktxt = booktxt[:sidx]
    
    booktxt = booktxt.strip()
    
    with open(OUTPUT_FOLDER + '/{}__{}.txt'.format('gutenberg', str(bookid)), 'w') as f:
        f.write(booktxt)
          
    print('Done')
```
